```python

frames = []
for i in range(0, len(data['data'])):
    A = data['data'][i]
    # reshape the data and append it to frames 
    data_csv = pd.DataFrame(np.array(A).reshape(-1,len(A)), columns=data['columns'])
    frames.append(data_csv)

```

```python
# concat the data 
result = pd.concat(frames)
result.head()    
```
