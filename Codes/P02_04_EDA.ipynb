````python
col_1_list = [ 'BDE? (e)', 'Qual. (g)','dV (km/s) (h)', 
               'B (nT) (k)', 'MC? (l)',
               'Dst (nT) (m)', 'Normalized_time_interval']

col_2_list = col_1_list
````

````python
# binary combinations

for col1 in col_1_list:
    for col2 in col_2_list:
#         print(col1)
        if col1 == col2:
            pass
        else:
            
            new_col_name = col1 + '*' + col2 + '**'
    #         col11 = col1.replace(' ','_')
    #         col22 = col2.replace(' ','_')
            df[new_col_name] = df[col1]*df[col2]
````

````python
df_corr = df.corr().abs()
plt.figure(figsize=(4,20))
sns.heatmap(df_corr[['V_max (km/s) (j)']].sort_values(by=['V_max (km/s) (j)'],ascending=False),
            vmin=0,
            cmap='Blues',
            annot=True);
# plt.savefig("../Figures/" + section_fig_main_title +"corr_3.png")
plt.show()
````

````python
col_3_list = col_1_list
````

````python
# binary combinations

for col1 in col_1_list:
    for col2 in col_2_list:
        for col3 in col_3_list:
        
    #         print(col1)
            if col1 == col2:
                pass
            else:

                new_col_name = col1 + '*' + col2 + '*' + col3 +'***'
        #         col11 = col1.replace(' ','_')
        #         col22 = col2.replace(' ','_')
                df[new_col_name] = df[col1]*df[col2]*df[col3]
````

````python
df_corr = df.corr().abs()
plt.figure(figsize=(4,100))
sns.heatmap(df_corr[['V_max (km/s) (j)']].sort_values(by=['V_max (km/s) (j)'],ascending=False),
            vmin=0,
            cmap='Blues',
            annot=True);
# plt.savefig("../Figures/" + section_fig_main_title +"corr_3.png")
plt.show()
````
