```python
import os, sys
import requests
from urllib.parse import urljoin, urlparse
from bs4 import BeautifulSoup
import re
```

## Request
Send request to the data source table webpage [here](http://www.srl.caltech.edu/ACE/ASC/DATA/level3/icmetable2.htm#(k))

```python
url = "http://www.srl.caltech.edu/ACE/ASC/DATA/level3/icmetable2.htm#(k)"
req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.contents
```

## Save the webpage html file

```python
with open("../Data/ICME_WebScr_data.html", "w") as f: 
    f.write(str(soup.contents))
```

## Read the data html from local drive

```python
# source https://www.geeksforgeeks.org/how-to-parse-local-html-file-in-python/
# Importing BeautifulSoup class from the bs4 module
from bs4 import BeautifulSoup

# Opening the html file
HTMLFile = open("../Data/ICME_WebScr_data.html", "r")

# Reading the file
index = HTMLFile.read()

# Creating a BeautifulSoup object and specifying the parser
S = BeautifulSoup(index, 'lxml')


# Using the prettify method to modify the code
print(S.body.prettify())
```
