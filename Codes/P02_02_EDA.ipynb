## BDE

Evidence of BiDirectional suprathermal Electron strahls (BDE) in ACE/SWEPAM Observations. If data are unavailable from SWEPAM (data commence on 10/22/97), observations from the 3-D P instrument on WIND are referred to. "SEP" indicates that an intense solar energetic particle event was in progress at the time of ICME passage and electron flows therefore cannot be determined.

````python
df['BDE? (e)'].value_counts().sort_index()
````

````python
df.groupby("BDE? (e)")["V_max (km/s) (j)"].describe().T
````

````python
# check out BDE as target

sub_BDE_3dot_mask = df["BDE? (e)"] == '...'
sub_BDE_N_mask = df["BDE? (e)"] == 'N'
sub_BDE_SEP_mask = df["BDE? (e)"] == 'SEP'
sub_BDE_Y_mask = df["BDE? (e)"] == 'Y'

plt.figure(figsize=(12,6))
plt.hist([df[sub_BDE_3dot_mask]['V_max (km/s) (j)'], 
          df[sub_BDE_N_mask]['V_max (km/s) (j)'],
          df[sub_BDE_SEP_mask]['V_max (km/s) (j)'], 
          df[sub_BDE_Y_mask]['V_max (km/s) (j)']],
         
         bins=20, color=['goldenrod', 'grey','black', 'red'], 
         ec='k', density=True)
plt.title('Normalized V_max (km/s) (j) distribution by \nBiDirectional suprathermal Electron strahls', fontsize=20)
plt.xlim(0,)
plt.ylabel("density"), 
plt.xlabel("V_max (km/s) (j)")
plt.legend(['No Data', 'N', 'An intense solar energetic particle \nevent was in progress at the time', 'Y'])


text = "Observations: unfortunately, there is no definition for N and Y presented on the webpage. \
        \nTherefore I just infer from a black color bar in the above figure. As it is seen, \
        \nthe coronal mass ejection speed measured at the maximum during the intense solar particle \
        \nevents makes sense. The solar particle events affect the measurement of solar winds \
        \nmeasurements apparatuses. "
plt.figtext(0.5, -0.16, text, 
            ha="center", fontsize=14, 
            bbox={"facecolor":"orange", 
                  "alpha":0.5, "pad":5})




plt.savefig("../Figures/" + section_fig_main_title +"BDE.png")

plt.show()
````

unfortunately, there is no definition for N and Y presented on the webpage. Therefore I just infer from a black color bar in the above figure. As it is seen, the coronal mass ejection speed measured at the maximum during the intense solar particle events makes sense. The solar particle events affect the measurement of solar winds measurements apparatuses. 

## MC

'2' indicates that a magnetic cloud has been reported in association with the ICME (see (d) above) or (occasionally, or for recent events) that by our assessment, the ICME has the clear features of a magnetic cloud but a magnetic cloud may not have been reported. 'H' indicates an event reported by Huttunen et al. Ann Geophys. (2005) 23:1-17 that is not listed by Lepping. '1' indicates that the ICME shows evidence of a rotation in field direction, but lacks some other characteristics of a magnetic cloud, for example an enhanced magnetic field. '0' indicates that the ICME is not a reported magnetic cloud, and lacks most of the typical features of a magnetic cloud, such as a smoothly rotating, enhanced magnetic field.

````python
df.groupby("MC? (l)")["V_max (km/s) (j)"].describe().T
````

````python
# check out MC? (l) as target

sub_MC_0_mask = df["MC? (l)"] == '0'
sub_MC_1_mask = df["MC? (l)"] == '1'
sub_MC_2_mask = df["MC? (l)"] == '2'
sub_MC_2H_mask = df["MC? (l)"] == '2H'

plt.figure(figsize=(12,6))
plt.hist([df[sub_MC_0_mask]['V_max (km/s) (j)'], 
          df[sub_MC_1_mask]['V_max (km/s) (j)'],
          df[sub_MC_2_mask]['V_max (km/s) (j)'], 
          df[sub_MC_2H_mask]['V_max (km/s) (j)']],
         
         bins=20, color=['goldenrod', 'grey','black', 'red'], 
         ec='k', density=True)
plt.title('Normalized V_max (km/s) (j) distribution by \nmagnetic cloud', fontsize=20)
plt.xlim(0,)
plt.ylabel("density"), 
plt.xlabel("V_max (km/s) (j)")
plt.legend(['ICME is not a reported magnetic cloud, and lacks \nmost of the typical features of a magnetic cloud', 
            'ICME shows evidence of a rotation in field \ndirection, but lacks some other characteristics', 
            'magnetic cloud has been reported in association \nwith the ICME ', 
            'ICME is not a reported magnetic cloud, and lacks \nmost of the typical features of a magnetic cloud \nevent reported by Huttunen et al'])




text = "Observations: from the above figure, It is clear that the events reported by Huttunen et al \
        \nhave reported some outliers in the overall data set. It is seen that the Huttunen et al group \
        \nhave reported some very high ICME speeds. This observation opens room for investigating the \
        \nwhole data reported by their team to discover the reason behind these outliers. is there any \
        \ndifference between the measurement tools used by their team?"
plt.figtext(0.5, -0.16, text, 
            ha="center", fontsize=14, 
            bbox={"facecolor":"orange", 
                  "alpha":0.5, "pad":5})







plt.savefig("../Figures/" + section_fig_main_title +"MC.png")

plt.show()
````

from the above figure, It is clear that the events reported by Huttunen et al have reported some outliers in the overall data set. It is seen that the Huttunen et al group have reported some very high ICME speeds. This observation opens room for investigating the whole data reported by their team to discover the reason behind these outliers. is there any difference between the measurement tools used by their team?

## Qual. (g)

The "quality" of the boundary times (`1' indicating the most reliable) based on assessment of the various data sets, including plasma, magnetic field and solar wind composition/charge states. 'W' indicates that the overall ICME signatures are particularly weak.

````python
df.groupby("Qual. (g)")["V_max (km/s) (j)"].describe().T
````

````python

sub_Qual_1_mask = df["Qual. (g)"] == '1'
sub_Qual_2_mask = df["Qual. (g)"] == '2'
sub_Qual_2W_mask = df["Qual. (g)"] == '2W'
sub_Qual_3_mask = df["Qual. (g)"] == '3'
sub_Qual_3W_mask = df["Qual. (g)"] == '3W'


plt.figure(figsize=(12,6))
plt.hist([df[sub_Qual_1_mask]['V_max (km/s) (j)'], 
          df[sub_Qual_2_mask]['V_max (km/s) (j)'],
          df[sub_Qual_2W_mask]['V_max (km/s) (j)'], 
          df[sub_Qual_3_mask]['V_max (km/s) (j)'],
         df[sub_Qual_3W_mask]['V_max (km/s) (j)']],
         
         bins=20, color=['goldenrod', 'grey','black', 'red', 'green'], 
         ec='k', density=True)
plt.title('Normalized V_max (km/s) (j) distribution by \nquality of the boundary times', fontsize=20)
plt.xlim(0,)
plt.ylabel("density"), 
plt.xlabel("V_max (km/s) (j)")
plt.legend(['most reliable', 
            'reliable', 
            'reliable - overall ICME signatures are particularly weak', 
            'less reliable',
           'less reliable - overall ICME signatures are particularly weak'])


text = "Observations: the data is mostly reliable although there are some in-reliable \n\
values for ICME speeds between the 300 km/s to 600 km/s."
plt.figtext(0.45, -0.05, text, 
            ha="center", fontsize=14, 
            bbox={"facecolor":"orange", 
                  "alpha":0.5, "pad":5})

plt.savefig("../Figures/" + section_fig_main_title +"Qual.png")

plt.show()
````
