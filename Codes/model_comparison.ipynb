```python
section_fig_main_title = 'model_comparison'
```

```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
with open("../Data/Multivar_tf_supervised_P03_06_performance.json", 'r') as f:
    Multivar_tf_supervised_P03_06_performance = json.load(f)
    
with open("../Data/ar_150lgs_P03_01_performance.json", 'r') as f:
    ar_150lgs_P03_01_performance = json.load(f)
    
with open("../Data/arima_510_P03_01_performance.json", 'r') as f:
    arima_510_P03_01_performance = json.load(f)   
    
with open("../Data/lr_60lgs_P03_04_performance.json", 'r') as f:
    lr_60lgs_P03_04_performance = json.load(f)
    
with open("../Data/lstm1_P03_02_performance.json", 'r') as f:
    lstm1_P03_02_performance = json.load(f)
    
with open("../Data/lstm2_P03_02_performance.json", 'r') as f:
    lstm2_P03_02_performance = json.load(f)
    
with open("../Data/lstm_P03_05_performance.json", 'r') as f:
    lstm_P03_05_performance = json.load(f)
    
with open("../Data/squen_60lgs_P03_04_performance.json", 'r') as f:
    squen_60lgs_P03_04_performance = json.load(f)
    
with open("../Data/xgboost_60lgs_P03_04_performance.json", 'r') as f:
    xgboost_60lgs_P03_04_performance = json.load(f)
```

```python
model_names = [
    'Multivariate_tf_supervised_DNN\nP03_06_performance',
    'Auto_Regressive_150lags\nP03_01_performance',
    'ARIMA_510\nP03_01_performance',
    'LR_60lags\nP03_04_performance',
    'LSTM_1\nP03_02_performance',
    'LSTM_2\nP03_02_performance',
    'LSTM\nP03_05_performance',
    'Sequential model_60lags\nP03_04_performance',
    'xgboost_60lags\nP03_04_performance']



performances_rsq = [
Multivar_tf_supervised_P03_06_performance['R_sq'],
ar_150lgs_P03_01_performance['R_sq'],
arima_510_P03_01_performance['R_sq'],
lr_60lgs_P03_04_performance['R_sq'],
lstm1_P03_02_performance['R_sq'],
lstm2_P03_02_performance['R_sq'],
lstm_P03_05_performance['R_sq'],
squen_60lgs_P03_04_performance['R_sq'],
xgboost_60lgs_P03_04_performance['R_sq']]

performances_rmse= [
Multivar_tf_supervised_P03_06_performance['rmse'],
ar_150lgs_P03_01_performance['rmse'],
arima_510_P03_01_performance['rmse'],
lr_60lgs_P03_04_performance['rmse'],
lstm1_P03_02_performance['rmse'],
lstm2_P03_02_performance['rmse'],
lstm_P03_05_performance['rmse'],
squen_60lgs_P03_04_performance['rmse'],
xgboost_60lgs_P03_04_performance['rmse']]
```

```python



plt.rcdefaults()
fig, ax = plt.subplots(figsize=(16,7))

# Example data
# people = model_names
y_pos = np.arange(len(model_names))
performance = performances_rsq
error = np.random.rand(len(model_names))


# color = []


ax.barh(y_pos, performance, align='center', color='maroon')
ax.set_yticks(y_pos)
ax.set_yticklabels(model_names)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('R-sq', fontsize=18)
ax.set_title('Models performance', fontsize=18)
plt.xticks(fontsize=16), plt.yticks(fontsize=16)

plt.xlim(0,)

fig.set_tight_layout(True)

plt.savefig("../Figures/" + section_fig_main_title + "model_performance_r2.png")
plt.show()
```

```python
plt.rcdefaults()
fig, ax = plt.subplots(figsize=(16,7))

# Example data
# people = model_names
y_pos = np.arange(len(model_names))
performance = performances_rmse
error = np.random.rand(len(model_names))

ax.barh(y_pos, performance, align='center', color='maroon')
ax.set_yticks(y_pos)
ax.set_yticklabels(model_names)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('RMSE', fontsize=18)
ax.set_title('Models performance', fontsize=18)
plt.xticks(fontsize=16), plt.yticks(fontsize=16)
fig.set_tight_layout(True)

plt.savefig("../Figures/" + section_fig_main_title + "model_performance_RMSE.png")

plt.show()
```
